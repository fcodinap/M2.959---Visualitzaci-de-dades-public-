# Data
---

```python
# Dades ficticies
sample_data = [
  {"brand": "Samsung", "share": 19, "color": "#004ea2"},
  {"brand": "Apple", "share": 18, "color": "#555555"},
  {"brand": "Xiaomi", "share": 13, "color": "#ff6700"},
  {"brand": "OPPO/vivo", "share": 18, "color": "#11a05b"},
  {"brand": "Others", "share": 32, "color": "#cccccc"}
]
```

# Entorn
---


```python
import pandas as pd
import matplotlib.pyplot as plt

# La llibreria no es troba a collab
!pip install pywaffle
from pywaffle import Waffle
```

# Visualització
---

```python
# Preparació Dades en dataframe
df = pd.DataFrame(sample_data)
data_dict = dict(zip(df.brand, df.share))

#Llegenda
custom_labels = [f"{row['brand']} ({row['share']}%)" for _, row in df.iterrows()]

# Plot
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    labels = custom_labels,
    values=data_dict,
    colors=list(df.color),
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    icons='circle',
    font_size=25,
    title={'label': 'Ventes Smartphones 2026', 'loc': 'center'},
    starting_location='NW',
    vertical='false'
)

# Descripcions
plt.figtext(
    0.12, -0.1,
    "Mostra grafica Dot Matrix o Waffle Chart",
    fontsize=12,
    color='#999999',
    style='italic'
)

plt.figtext(
    0.12, -0.15,
    "Percentatges de ventes al mercat per marca de Telefon",
    fontsize=12,
    color='#999999',
    style='oblique'
)

# Visualitzacio
plt.show()
```
